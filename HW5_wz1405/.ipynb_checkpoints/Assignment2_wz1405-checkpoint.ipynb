```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q=SELECT' 

import urllib2
import urllib
import StringIO
import ast
import pandas as pd
try:
    import urllib2 as ulib
except ImportError:
    import urllib3 as ulib

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError

import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

```python
API_URL = 'https://fb55.carto.com/api/v2/sql?q=SELECT'

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
# 1.1 sort data by both: start_station_id ascending, and tripduration descending 
# Ref: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html
# • Sort data by start_station_id, tripduration
# - Only checking trips with duration <= 3 hours 

# Ref: https://www.w3schools.com/sql/sql_orderby.asp
query = '''SELECT * FROM citibike
WHERE tripduration<=10800
ORDER BY start_station_id ASC
;
'''

table = get_data(query)
print table.head()

query = '''SELECT * FROM citibike
WHERE tripduration<=10800
ORDER BY tripduration DESC
;
'''

table = get_data(query)
print table.head()
```

```python
# 3.1 Showing the list of start station locations 
# hint: GROUP BY station id, COUNT();

query = ''' SELECT start_station_name FROM citibike 
            GROUP BY start_station_name
            ;'''
table = get_data(query)
table.head()
```

```python
#3.2 Showing the number of trips started per station 
# Ref: https://www.w3schools.com/sql/sql_groupby.asp
query = ''' SELECT start_station_name, COUNT(cartodb_id) FROM citibike
            GROUP BY start_station_name    
            ORDER BY COUNT(cartodb_id) DESC
            ;'''
NumberOfTrips = get_data(query)
print NumberOfTrips.head()
```

```python
# 3.3  but only for stations within 500m of Time Square!
# - The coordinates of Time Square is (40.7577,-73.9857)
# Ref: dexxtr.com

query = ''' SELECT start_station_name, COUNT(cartodb_id) FROM citibike
            GROUP BY start_station_name    
            ORDER BY COUNT(cartodb_id) DESC
            ;'''
NumberOfTrips = get_data(query)
print NumberOfTrips.head()

query = ''' SELECT start_station_name, start_station_latitude, start_station_longitude FROM citibike


WHERE(SQRT(POW(start_station_latitude-40.7577 , 2) + POW(start_station_longitude - -73.9857 , 2)) * 111000 < 500)

;'''


table = get_data(query)
print table.head()
```

```python
#3.3 
query = ''' SELECT start_station_name, COUNT(cartodb_id) FROM citibike
            WHERE(SQRT(POW(start_station_latitude-40.7577 , 2) + POW(start_station_longitude - -73.9857 , 2)) * 111000 < 500)
            GROUP BY start_station_name    
            ORDER BY COUNT(cartodb_id) DESC
            ;'''
NumberOfTrips = get_data(query)
print NumberOfTrips
```

```python
#4.1 Find the station that had the 1. longest average trip duration 2. during weekends and 3. within 500m of TimeSquare! 


query = ''' SELECT start_station_name, starttime, MAX(tripduration) as TripDuration, (SQRT(POW(start_station_latitude - 40.7577 , 2) + POW(start_station_longitude - -73.9857 , 2)) * 111000 < 500) as Radius FROM citibike
            
            WHERE (SQRT(POW(start_station_latitude - 40.7577 , 2) + POW(start_station_longitude - -73.9857 , 2)) * 111000 < 500)

            and EXTRACT(dow from starttime) in (0, 6)
            GROUP BY start_station_name, starttime, TripDuration, Radius
            ORDER BY tripduration DESC
            LIMIT 1

;
'''
table = get_data(query)
print table.head()

# Group BY must include every term mentioned in Select in the first sentence. 
```

```python
#4.2 Extra: create lines for trips started from stations within 500m of Times
#Squares and lasted less than 2 hours. The number of trips per each
#pair of stations are output as attributes of these lines.
query = ''' SELECT start_station_name, starttime, MAX(tripduration) as TripDuration, (SQRT(POW(start_station_latitude - 40.7577 , 2) + POW(start_station_longitude - -73.9857 , 2)) * 111000 < 500) as Radius FROM citibike
            
            WHERE (SQRT(POW(start_station_latitude - 40.7577 , 2) + POW(start_station_longitude - -73.9857 , 2)) * 111000 < 500)

            and EXTRACT(dow from starttime) in (0, 6)
            and tripduration<=7200
            GROUP BY start_station_name, starttime, TripDuration, Radius
            ORDER BY tripduration DESC

;
'''
table = get_data(query)
print table.head()
```
